## matplotlib

 - 2차원 그래픽 패키지
 - 파이썬에서 데이터를 이용해 그래프를 그려주는 가장 기본적인 라이브러리 패키지
 - figure라는 그림 단위를 사용하며 이 안에서 하나 혹은 여러개의 그래프를 그리고 관리할 수 있다.
 - 이 때 figure 안에 들어가는 plot 하나를 subplot이라고 한다. 
 - line plot, bar chart, pie chart, histogram, box plot, scatter plot 다양한 스타일을 지원
 - 특히, 이번에는 matplotlib 라이브러리의 pyplot 모듈을 사용한다.

```python
import matplotlib.pyplot as plt # matplotlib를 불러온다.
%matplotlib inline  
# cell magic 명령어 / 그래프의 결과를 출력 세션에 나타나게 하는 설정 / 파이썬 코드가 아닌 쥬피터노트북에 특정 기능을 수행하게 하는 명령어

# 그 외 필요 모듈 불러오기
import pandas as pd
import numpy as np
import os
```

```python
x = np.arange(0, 12, 0.01)
y = np.sin(x)

plt.figure(figsize = (10,6))
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.plot(x,y)
plt.grid() #격자무늬를 넣으라는 명령
plt.xlabel('Time') # x축 라벨 정하기
plt.ylabel('Amplitude') # y축 라벨 정하기
plt.title('Example of sin wave') # 그래프 타이틀 정하기
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x)) # 플랏 명령을 두개 넣으면 그래프 2개가 동시에 보인다.
plt.grid() #격자무늬를 넣으라는 명령
plt.xlabel('Time') # x축 라벨 정하기
plt.ylabel('Amplitude') # y축 라벨 정하기
plt.title('Example of sin wave') # 그래프 타이틀 정하기
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.plot(x,np.sin(x), label = 'sin') # label 옵션으로 범례(legend)를 추가한다.
plt.plot(x,np.cos(x), label = 'cos')
plt.grid() #격자무늬를 넣으라는 명령
plt.legend() # 범례 추가
plt.xlabel('Time') # x축 라벨 정하기
plt.ylabel('Amplitude') # y축 라벨 정하기
plt.title('Example of sin wave') # 그래프 타이틀 정하기
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.plot(x,np.sin(x), label = 'sin', lw = 3) # lw 옵션으로 선의 굵기를 지정한다.
plt.plot(x,np.cos(x), label = 'cos', color = 'r') # color 옵션으로 그래프의 색을 지정한다.
plt.grid() #격자무늬를 넣으라는 명령
plt.legend() # 범례 추가
plt.xlabel('Time') # x축 라벨 정하기
plt.ylabel('Amplitude') # y축 라벨 정하기
plt.title('Example of sin wave') # 그래프 타이틀 정하기
plt.show()
```

## 데이터 병합하기

```python
df1 = pd.DataFrame({'Year' : [1992, 1993, 1989, 1994],
                   'Height' : [163, 155, 153, 180],
                   'Name' : ['a', 'b', 'c', 'd']},
                  index = [1,2,3,4])
df2 = pd.DataFrame({'Year' : [1992, 1993, 1989, 1994],
                   'Height' : [163, 155, 153, 180],
                   'Name' : ['a', 'b', 'c', 'd']},
                  index = [5,6,7,8])
df3 = pd.DataFrame({'Year' : [1992, 1993, 1989, 1994],
                   'Height' : [163, 155, 153, 180],
                   'Age' : [28, 27, 31, 26]},
                  index = [1,2,3,4])
```

```python
pd.concat([df1, df2]) # 아래에 붙는다
```

```python
pd.concat([df1, df3]) # 공통 부분이 없는 것은 NaN처리
```

```python
df1.append(df2) # 아래에 붙는다.
```

```python
df1.append(df3) # 공통 부분이 없는 것은 NaN처리
```

```python
pd.merge(df1, df2) # 공통된 부분만 붙는다.
```

```python
pd.merge(df1, df2, on = 'Year') # 따로 기준을 걸어주면 공통부분은 합쳐지지 않고 중복 되었다고 표시됨
```

```python
pd.merge(df1, df2, on = ['Year', 'Name']) # 기준을 2개로 할 수 있음
```

```python
pd.merge(df1, df3, how = 'outer')
```

```python
pd.merge(df1, df3, on = 'Year' , how = 'inner')
```

## if, for statement in one line

```python
my_list = [2,3,4,5]
new_list = []

for i in my_list : 
    new_list.append(i*2)

new_list
```

```python
my_list = [2,3,4,5]

new_list = list([i*2 for i in my_list])
new_list
```

```python
my_list = [1,2,3,4,5]
new_list = []

for i in my_list : 
    if i > 3 : 
        new_list.append(i*2)
    else : 
        new_list.append(i)
        
new_list
```

```python
list([i*2 if i > 3 else i for i in my_list])
```

```python
my_list = [2,3,4,5]
new_list = []

for i in my_list : 
    if i > 3 : 
        new_list.append(i*2)
        
new_list
```

```python
list([i*2 for i in my_list if i > 3])
```

### 기본 피벗 테이블

- 중복 index는 하나로 합쳐진다.
- 특별히 지정하지 않았다면 숫자형 데이터 컬럼만 남고 value는 평균을 나타낸다.

```python
pd.pivot_table(df, index = ['Name'])
```

### index 여러개 지정

```python
pd.pivot_table(df, index = ['Name', 'Rep', 'Manager'])
```

### 특정 value 지정

```python
pd.pivot_table(df, index = ['Manager', 'Rep'], values = ['Price'])
```

### value 반환 값 바꾸기 (sum)

```python
pd.pivot_table(df, index = ['Manager', 'Rep'], values = ['Price'], aggfunc = np.sum)
```

```python
pd.pivot_table(df, index = ['Manager', 'Rep', 'Product'], values = ['Price', 'Quantity'],
              aggfunc = [np.sum, np.mean], fill_value = 0, margins = True) # 빈칸은 NaN이 아닌 0으로 채운다.
```
